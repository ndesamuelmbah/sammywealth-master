# We will first do a word cloud on the title
## Clean the title by:
<ol><li> <h4> Remove all punctuation marks. <h4/> 
    <li> <h4> Remove all english stop words </h4><ol/>
 

```python
from string import punctuation
from nltk.corpus import stopwords
my_stopwords = stopwords.words('english')
punctuation
```

```python
import re
clean_text = ' '.join([word for word in re.sub('[!"#$%&\'()*+,-./:”;<=>?@[\\]^_“`{|}~]', '', text.lower()).split() if not word in my_stopwords])
clean_text
```

### Import the library spacy and use it to lemmantize each word in out text.  
<p>If you need to install spacy, you may do so using</p>
<p> `pip install -U spacy` </p>
<p> If you are running into any problems, please checkout [spacy documentation here.](https://spacy.io/usage/#section-instructions)

```python
import spacy
nlp = spacy.load('en')
```

```python
lemmatized_text = ' '.join([word.lemma_ for word in nlp(clean_text)])
lemmatized_text
```

### Finally, import your wordcloud library and draw your wordcloud
If you do not have wordclould installed on you system, you may want to pip install it as shown below.
<br> <p> `pip install wordcloud` </p></br><p>
and checkout the [wordclould github repo](https://github.com/amueller/word_cloud) if you are having any issues.

```python
from wordcloud import WordCloud
wordcloud = WordCloud(font_path=None, width=400, height=200, margin=1, ranks_only=None, prefer_horizontal=0.9, mask=None, 
                      scale=1, color_func=None, max_words=2000, min_font_size=4, stopwords=None, random_state=None, 
                      background_color=None, max_font_size=50, font_step=1, mode='RGBA', relative_scaling='auto', 
                      regexp=None, collocations=True, colormap='viridis', normalize_plurals=True, contour_width=0, 
                      contour_color='black', repeat=False).generate(lemmatized_text)
```

```python
plt.figure(figsize=(15,10))
plt.imshow(wordcloud.recolor(random_state=17))
plt.title('WordCloud of Words in this page up to the paragraph below', fontsize = 30, color = 'red')
plt.axis("off")
```
